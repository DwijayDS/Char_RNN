Cleaning and merging Data

```python
'''Import libraries'''

import numpy as np
import pandas as pd
import nltk
from clean_text import clean_sent
import wandb
```

```python
'''Read Harry Potter Books and remove advertisement'''
harry=[]
for i in range(1,8):
    path='harry_txt/harry'+str(i)+'.txt'
    temp=open(path,encoding='utf-8',errors='ignore').read().split('\n')
    for i in temp:
        if len(i.split())>5 and i!="Generated by ABC Amber LIT Converter, http://www.processtext.com/abclit.html":
            harry.append(i)
del temp ,path,i 
```

```python
textfile = open("harry_txt/harry_data.txt", "w")
for element in harry:

    textfile.write(element + "\n")

textfile.close()
```

```python
'''
Cleaning text
1. Convert everything to lower
2. Remove special characters
3. Replace abreviations/ shotforms with actual words
4. Remove words with greater than 12 characters.

'''
clean_harry=[]
for i in harry:
    clean_harry.append(clean_sent(i))   
del i,harry
```

```python
textfile = open("harry_txt/cleaned_harry_data.txt", "w")
for element in clean_harry:

    textfile.write(element + "\n")

textfile.close()
```
